```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a pandas Dataframe
```

```python
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
credit_card_data['Class'].value_counts()
```

```python
# this dataset is highly unballance
```

```python
# seprating the data for analysis
```

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
print(legit)
```

```python
#statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the value for both the transaction
```

```python
credit_card_data.groupby('Class').mean()
```

```python
# Now we have to deal with the unballance data
# Under-Sampling
# build a sample dataset containing similar distribution and fraudlent transaction
```

```python
legit_sample = legit.sample(n=492)
```

```python
# concatenate legit+fraud
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail(5)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Model Training
```

```python
model = LogisticRegression()
```

```python
#training the logistic Regression with Training Data
```

```python
model.fit(X_train, Y_train)
```

```python
#Evaluation
```

```python
# accuracy on traning data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
